## Problem :  
### Duplicate Values

Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

```python
# Solution

def containsDuplicate(nums):

  if len(set(nums)) == len(nums):
    return False
  else:
    return True
```

Using Set() built in datatype of Python for more optimal solution which is O(N) rather than O(N^2) of nested loops.

## Problem :
### Missing Number
Given an array nums containing n distinct numbers in the range [0, n], return the only number in the range that is missing from the array.

```python
# Solution 1:
def missing_number(nums):
  nums.sort()

  for i,v in enumerate(nums):
    if i =! v:
      return v+1
    elif v == len(nums)-1:
      return v+1

# this solution is O(Nlogn), the reason is that sorting in python is nlogn.
```

```python
# Solution 2:
def missing_number(nums):
  return sum(range(len(nums)+1)) - len(nums)

# O(N) : Iterate through list and sum (twice: once for given list and once for expected using range)
# Thats why this solution is O(n)
```

## Problem :
### Find All Missing Numbers in an array
Given an array nums of n integers where nums[i] is in the range [1, n], return an array of all the integers in the range [1, n] that do not appear in nums.

```python
# Solution

def find_all_missing_num(nums):

    set_nums = set(nums)
    ret = []
    for i in range(1,len(nums) + 1):
      if i not in set_nums:
        ret.append(i)
      return ret

# this solution is O(N), as we are iterating over an array of len(n)
```

## Problem :
### Two Sum
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

```python
# Solution 1
def two_sum(nums, target):
  for i in range(len(nums) + 1):
    for j in range(i + 1, len(nums)):
      if nums[i] + nums[j] == target:
        return [i,j]

# This soultion is O(N^2), n*n, n lenght of list and n for each number of list
```

```python
two_sum([4,6,7,8,8,3,2],10)
```

#### Intuition for this Solution

Hash Map is pythons Dict

We can use a hash map to solve this problem. First, we initialize an empty hash map. Then, we loop through the nums array, checking and saving each index and value in the hash map. For each value, we subtract it from the target value and check if the resultant value is in the hash map. If it is, we return the current index and the index of the value that is stored in the hash map. Otherwise, we save the current value in the hash map.

```python
# Solution 2

def two_sum(nums, target):
  hash_map = {}
  for i,v  in enumerate(len(nums) + 1):
    diff = target-v
    if diff in hash_map:
      return [i, hash_map[diff]]
    hash_map[v] = i

# This soultion in O(N) bcz it is searching hashmap which constant linear and looping thorugh a list.
```
